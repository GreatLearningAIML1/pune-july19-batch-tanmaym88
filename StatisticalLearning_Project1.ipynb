```python
laliga = pd.read_csv('Laliga.csv')
```

```python
laliga.replace('-',0,inplace=True)
```

```python
laliga[(laliga['Debut'].apply(lambda x: int(x[0:4])) > 1930) & (laliga['Debut'].apply(lambda x: int(x[0:4])) < 1980)]['Team']
```

```python
laliga['Points'] = laliga['Points'].astype(str).astype(int)
laliga.sort_values(by='Points',ascending=False).iloc[0:5][['Team','Points']]
```

```python
laliga['GoalsFor'] = laliga['GoalsFor'].astype(str).astype(int)
laliga['GoalsAgainst'] = laliga['GoalsAgainst'].astype(str).astype(int)
def Goal_diff_count(df):
    df['Goal_diff_count'] = df['GoalsFor'] - df['GoalsAgainst']
    return df
laliga_with_goal_diff = Goal_diff_count(laliga)
```

```python
laliga_with_goal_diff.iloc[pd.Series.idxmax(laliga_with_goal_diff['Goal_diff_count']),:]['Team']
```

```python
laliga_with_goal_diff.iloc[pd.Series.idxmin(laliga_with_goal_diff['Goal_diff_count']),:]['Team']
```

```python
laliga['GamesWon'] = laliga['GamesWon'].astype(str).astype(int)
laliga['GamesPlayed'] = laliga['GamesPlayed'].astype(str).astype(int)
laliga['Winning Percent'] = laliga['GamesWon']/laliga['GamesPlayed']*100
laliga['Winning Percent'].fillna(0,inplace=True)
laliga.sort_values(by='Winning Percent',ascending=False).iloc[0:5][['Team','Winning Percent']]
```
